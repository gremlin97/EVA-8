```python
from datasets import load_dataset
eli5 = load_dataset('eli5_category', split = 'train[0:5000]')
```

```python
eli5 = eli5.train_test_split(test_size=0.2)
```

```python
eli5['train'][0]['answers']['text']
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained('distilbert/distilgpt2')
```

```python
eli5 = eli5.flatten()
```

```python
def preprocess_fx(entry):
  return tokenizer([''.join(x) for x in entry['answers.text']])

tokenized_eli5 = eli5.map(
    preprocess_fx,
    batched = True,
    num_proc = 4,
    remove_columns = eli5["train"].column_names
)
```

```python
tokenized_eli5['train']['input_ids'][1]
```

```python
block_size = 128


def group_texts(examples):
    # Concatenate all texts.
    concatenated_examples = {k: sum(examples[k], []) for k in examples.keys()}
    total_length = len(concatenated_examples[list(examples.keys())[0]])
    # We drop the small remainder, we could add padding if the model supported it instead of this drop, you can
    # customize this part to your needs.
    if total_length >= block_size:
        total_length = (total_length // block_size) * block_size
    # Split by chunks of block_size.
    result = {
        k: [t[i : i + block_size] for i in range(0, total_length, block_size)]
        for k, t in concatenated_examples.items()
    }
    result["labels"] = result["input_ids"].copy()
    return result
```

```python
lm_dataset = tokenized_eli5.map(group_texts, batched=True, num_proc=4)
```

```python
from transformers import DataCollatorForLanguageModeling

tokenizer.pad_token = tokenizer.eos_token
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
```

```python
from transformers import AutoModelForCausalLM, TrainingArguments, Trainer
model = AutoModelForCausalLM.from_pretrained("distilbert/distilgpt2")
```

```python
training_args = TrainingArguments(
    output_dir="eli5_distilgpt",
    evaluation_strategy="epoch",
    learning_rate=2e-5, #num_train_epochs for number of epochs, default is 3
    weight_decay=0.01,
    push_to_hub=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=lm_dataset["train"],
    eval_dataset=lm_dataset["test"],
    data_collator=data_collator,
)

trainer.train()
```

```python
import math
eval_results = trainer.evaluate()
print(f"Perplexity: {math.exp(eval_results['eval_loss']):.2f}")
```
